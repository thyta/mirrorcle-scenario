```python
# # Define a function to extract Vietnamese phone numbers
# def extract_phone_numbers(text):
#     # Regex pattern for Vietnamese phone numbers
#     pattern = r'\b(0[3|5|7|8|9]\d{8})\b'
#     # Find all matches
#     matches = re.findall(pattern, text)
#     # Join all found numbers (if multiple numbers in the text)
#     return ', '.join(matches) if matches else None
def extract_phone_numbers(text):
    # Convert any non-string value to a string
    if not isinstance(text, str):
        text = str(text)
    
    # Regex pattern for Vietnamese phone numbers
    pattern = r'\b(0[3|5|7|8|9]\d{8})\b'
    # Find all matches
    matches = re.findall(pattern, text)
    # Join all found numbers (if multiple numbers in the text)
    return ', '.join(matches) if matches else None
```

```python
def doit(df):
    # Create a new column for phone numbers
    df['phone_numbers 1'] = df['Nội dung xử lý'].apply(extract_phone_numbers)
    df['phone_numbers 2'] = df['Nội dung tiếp nhận'].apply(extract_phone_numbers)
    # Insert 'phone_numbers' as the first column
    df.insert(0, 'phone_numbers 1', df.pop('phone_numbers 1'))
    # df.insert(1, 'phone_numbers 2', df.pop('phone_numbers 2'))
    return df
```

```python
doit(df)
```
